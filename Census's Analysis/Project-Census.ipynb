```python
rbabies=(babiesper100k2529/babiesper100k3034)*100
print("rate of birth increased by %f in 5 years from 25-29 to 30-34" %round(rbabies,2))
```

```python
people5660=df[(age>=56) & (age<=60)]
people6165=df[(age>=61)&(age<=65)]
deaths=len(people5660)-len(people6165)
tot=deaths*mag
print("Age group 56-60 : %f and Age group 61-65 : %f "% (len(people5660),(len(people6165))))
print("Deaths are: %d"%(deaths))
print("death rate of 100K is of 5 years : %f "%(tot))
print("Death Rate per annum per 100,000 people is : %f"%((deaths*mag)/5))
```

```python
people6670=df[(age>=66) & (age<=70)]
people7175=df[(age>=71) & (age<=75)]
deaths1=len(people6670)-len(people7175)
tot1=deaths1*mag
print("Age group 66-70 : %f and Age group 71-75 : %f "% (len(people6670),(len(people7175))))
print("Deaths are: %d"%(deaths1))
print("death rate of 100K is of 5 years : %f "%(tot1))
print("Death Rate per annum per 100,000 people is : %f"%((deaths*mag)/5))
```

```python
people7680=df[(age>=76) & (age<=80)]
people8085=df[(age>=81) & (age<=85)]
deaths2=len(people7680)-len(people8085)
tot2=deaths2*mag
print("Age group 76-80 : %f and Age group 81-85 : %f "% (len(people7680),(len(people8085))))
print("Deaths are: %d"%(deaths2))
print("death rate of 100K is of 5 years : %f "%(tot2))
print("Death Rate per annum per 100,000 people is : %f"%((deaths*mag)/5))
```

```python
tot_deaths=(tot+tot1)
print(int(tot_deaths))
tot_bab=babiesper100k2529+babiesper100k3034
print(int(tot_bab))
```

```python
diff_bd=int(tot_deaths-tot_bab)
if tot_deaths>tot_bab:
    print("Population is decreasing by %d"%diff_bd)
else:
    print("population is increasing by %d"%diff_bd)
```
