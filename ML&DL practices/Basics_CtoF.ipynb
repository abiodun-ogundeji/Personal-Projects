```python
import tensorflow as tf
tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)

import numpy as np
```

```python
celsius_q    = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit_a = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)

for i,c in enumerate(celsius_q):
  print("{} degrees Celsius = {} degrees Fahrenheit".format(c, fahrenheit_a[i]))
```

```python
l0 = tf.keras.layers.Dense(units=1, input_shape=[1])
```

```python
model = tf.keras.Sequential([l0])
```

Note: you would often see the layers defined inside the model definition

model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])

```python
model.compile(loss='mean_squared_error', 
              optimizer=tf.keras.optimizers.Adam(0.1))
```

```python
history = model.fit(celsius_q, fahrenheit_a, epochs=500, verbose=False)
print("Finished training the model")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("Epoch Number")
plt.ylabel("Loss Magnitude")
plt.plot(history.history['loss'])
```

```python
print(model.predict([100.0]))
```

```python
print("These are the layer variable: {}".format(l0.get_weights()))
```

A little experiment

```python
l0 = tf.keras.layers.Dense(units=4, input_shape=[1])
l1 = tf.keras.layers.Dense(units=4)
l2 = tf.keras.layers.Dense(units=4)
model = tf.keras.Sequential([l0, l1, l2])
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.1))
model.fit(celsius_q, fahrenheit_a, epochs=500, verbose=False)
print("Finished training the model")
print(model.predict([100.0]))
print("These are the l0 variable: {}".format(l0.get_weights()))
print("These are the l1 variable: {}".format(l1.get_weights()))
print("These are the l2 variable: {}".format(l2.get_weights()))
```
